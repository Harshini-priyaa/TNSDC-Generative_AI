```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/cancer dataset/cancer_reg.csv")

# Count the occurrences of values in the 'target_deathrate' column
death_rate_counts = df['target_deathrate'].value_counts()
print("Count of death rate:")
print(death_rate_counts)
```

```python
# Drop rows where a certain condition is met, for example, where 'avgdeathsperyear' is less than 100
df = df[df['avgdeathsperyear'] >= 100]
```

```python
# Find the sum of duplicated rows
duplicated_sum = df.duplicated().sum()
print("Sum of duplicated values:", duplicated_sum)
```

```python
# Sum of null values for each column
null_sum = df.isnull().sum()
print("Sum of null values:")
print(null_sum)
```

```python
import matplotlib.pyplot as plt

# Plot histogram for 'target_deathrate'
plt.hist(df['target_deathrate'], bins=20)
plt.xlabel('Death Rate')
plt.ylabel('Frequency')
plt.title('Histogram of Death Rate')
plt.show()
```

```python
# Pie chart for 'target_deathrate' and 'avgdeathsperyear'
plt.figure(figsize=(8, 8))
labels = ['Death Rate', 'Average Deaths']
sizes = [df['target_deathrate'].mean(), df['avgdeathsperyear'].mean()]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart of Death Rate vs Average Deaths')
plt.show()
```

```python
# Boxplot for 'avgdeathsperyear'
plt.boxplot(df['avgdeathsperyear'])
plt.ylabel('Average Deaths per Year')
plt.title('Boxplot of Average Deaths per Year')
plt.show()
```

```python
plt.bar(df['povertypercent'].index, df['povertypercent'].values)
plt.xlabel('Index')
plt.ylabel('Poverty Percent')
plt.title(' Graph of Poverty Percent')
plt.show()
```
